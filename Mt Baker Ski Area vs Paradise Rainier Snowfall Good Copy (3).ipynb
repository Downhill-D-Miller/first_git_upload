```python
def create_double(t, w, n, d):
    return[t*x + w*n for x in range(d)]

baker_x = create_double(2, 0.8, 1, 5)
paradise_x = create_double(2, 0.8, 2, 5)

plt.figure(figsize = (10, 6))
plt.bar(baker_x, baker_by_decade_averages, color = 'green')
plt.bar(paradise_x, paradise_by_decade_averages, color = 'gold')
middle_x = [(a+b) / 2.0 for a, b in zip(baker_x, paradise_x)]
ax = plt.subplot()
ax.set_xticks(middle_x)
ax.set_xticklabels(decades)
plt.legend(['Baker', 'Paradise'])
plt.title('Baker Vs Paradise Snowfall')
plt.xlabel('Decade')
plt.ylabel('Snowfall (inches)')
plt.show()
```

```python
def freq_tables(dataset):
    frequency_table = {'Low': 0, 'Medium': 0, 'High': 0, 'Ultimate': 0}
    prop_table = {}
    percent_table = {}
    for data in dataset:
        if data < 400:
            frequency_table['Low'] += 1
        elif 400 <= data < 600:
            frequency_table['Medium'] += 1
        elif 600 <= data < 800:
            frequency_table['High'] += 1
        elif data >= 800:
            frequency_table['Ultimate'] +=1
    for values in frequency_table:
        proportion = frequency_table[values] / len(dataset)
        percent = proportion * 100
        prop_table[values] = proportion
        percent_table[values] = percent
    return[frequency_table, prop_table, percent_table]
        
    
```

```python
baker_tables = freq_tables(baker_50_year_snowfall)
print(baker_tables)
baker_table_values = baker_tables[0].values()
baker_table_keys = baker_tables[0].keys()
```

```python
plt.figure(figsize = (10, 10))
plt.pie(baker_table_values, labels = baker_table_keys, autopct = '%.1d%%', colors = ['seagreen', 'darkorchid', 'indianred', 'gold'])
plt.title('Fifty-Year Mount Baker Snowfall History')
plt.show()
plt.savefig('Pie Chart Baker.png')
```
